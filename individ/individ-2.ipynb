В книжном магазине проводится анализ по продажам. Имеются данные по кол-ву проданных эпических и драматических произведений. Нужно сравнить их продажи за последний месяц и определить, какой жанр литературы больше нравится покупателям Найти среднее количество проданных книг каждого жанра.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# данные по продажам книг за месяц
epic_book = [60, 150, 120, 90, 110]
drama_book = [130, 120, 110, 100, 90]
```

```python
sr_ep = sum(epic_book) / len(epic_book)
sr_dr = sum(drama_book) / len(drama_book)

print("Среднее кол-во проданных эпических произведений:", sr_ep)
print("Среднее кол-во проданных драматических произведений:", sr_dr)
```

```python
labels = ['Неделя 1', 'Неделя 2', 'Неделя 3', 'Неделя 4', 'Неделя 5']

x = np.arange(len(labels))
width = 0.3

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, epic_book, width, label='Эпические')
rects2 = ax.bar(x + width/2, drama_book, width, label='драма')

ax.set_ylabel('Количество проданных книг')
ax.set_xlabel('Недели')
ax.set_title('Продажи книг за месяц')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```

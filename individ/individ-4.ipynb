Найти какое-либо изображение в сети Интернет. Создать ноутбук, в котором будет
отображено выбранное изображение средствами библиотеки matplotlib по URL из сети
Интернет.

```python
import matplotlib.pyplot as plt

from PIL import Image
import requests

from io import BytesIO


response = requests.get('https://koshka.top/uploads/posts/2021-12/thumbs/1638771511_1-koshka-top-p-milogo-kotika-v-shapochke-1.jpg')
img = Image.open(BytesIO(response.content))

plt.imshow(img)
```

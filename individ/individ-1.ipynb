Из орудия вылетает снаряд со скоростью v0 под углом α к горизонту. Определить: 
а) скорость (модуль и направление) и положение (координаты) снаряда в любой момент времени; 
б) время подъема до наивысшей точки и время полета;
в) высоту подъема и дальность полета.

```python
import math
import matplotlib.pyplot as plt
```

```python
# начальные данные
v0 = 50 
alpha = math.radians(30) 
g = 9.81 

# вычисляем время полета и время подъема
t_max = v0*math.sin(alpha)/g
t_flight = 2*t_max

# создаем списки для координат и скоростей
x_list = []
y_list = []
vx_list = []
vy_list = []

# вычисляем координаты и скорости снаряда в каждый момент времени
for t in range(int(t_flight*100)): 
    x = v0*math.cos(alpha)*t/100
    y = v0*math.sin(alpha)*t/100 - g*(t/100)**2/2
    vx = v0*math.cos(alpha)
    vy = v0*math.sin(alpha) - g*t/100
    x_list.append(x)
    y_list.append(y)
    vx_list.append(vx)
    vy_list.append(vy)

    # если снаряд упал на землю, прерываем цикл
    if y < 0:
        break
```

```python
# выводим результаты
print("Координаты снаряда:")
print("x =", x_list[-1], "м")
print("y =", y_list[-1], "м")
print("Скорости снаряда:")
print("vx =", vx_list[-1], "м/с")
print("vy =", vy_list[-1], "м/с")
```

```python
# рисуем график
plt.plot(x_list, y_list)
plt.title("Движение снаряда под углом к горизонту")
plt.xlabel("x, м")
plt.ylabel("y, м")
plt.show()
```
